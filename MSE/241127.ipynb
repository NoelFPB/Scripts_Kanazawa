```python
head = ['Loop','th','time']
# weight : 0 ~ 34
# iris data : 36,37,38,39

with open("D:\\[test]\\241122\\Ptsest.csv", 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(head)

    for loop in range(10):

        print("LOOP",loop,"------------------------")
        train, test = train_test_split(Id, test_size=0.3)       # 打乱数据集编号
        weight = np.round(np.random.uniform(0, 5, 35), 2)      # 生成weight部分的随机电压
        # 将weight输入
        voltage = ""
        for heater in range(35):
            voltage += f"{heater},{weight[heater]};"
        voltage_send = voltage +'\n'
        ser.write(voltage_send.encode())
        time.sleep(2)
            
        for th in train:
            # 将iris参数输入
            iris_send = f"{36},{SeL[th]};{37},{SeW[th]};{38},{PeL[th]};{39},{PeW[th]};" + '\n'
            start = time.time()
            ser.write(iris_send.encode())
            time.sleep(2)
            end = time.time()
            time_need = round((end-start),2)
            print(time_need)
            csvwriter.writerow([loop,th,time_need])
```
